```python
L = 100 # Length of lattice

basis = np.zeros((L,L))
i = 0
for v in basis:
    v[i] = 1
    i += 1

J = 0.5  # hopping strength
beta = (1 + np.sqrt(5))/2 # periodicity
phi = 0 # phase
lam = 2*J # disorder strength
U = 0
pbc = False # boundary condition (true = Periodic PBC, false = Open OBC)
mu = lam*np.cos(2*np.pi*beta*np.arange(L) + phi)    


def eigens(beta):
    mu = lam*np.cos(2*np.pi*beta*np.arange(L) + phi)

    ham = np.zeros((L,L))
    for i in range(L):
        ham = ham + -J*(np.outer(basis[i%L],basis[(i+1)%L].T)+np.outer(basis[(i+1)%L],basis[i%L])) + mu[i]*np.outer(basis[i], basis[i])

    return np.linalg.eigvals(ham)
```

```python
n = 200
beta = np.linspace(0,np.pi, n)

beta_plot = []
for i in range(n):
    beta_plot.append(np.ones(L)*beta[i])
    plt.plot(beta_plot[i], eigens(beta[i]), 'ro', markersize = 0.5)

plt.show()
    
```
